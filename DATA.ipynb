```python
data_df = data[data['fraudulent']==1]
```

```python
data_df
```

```python
country_codes_df = data['location'].str.split(',', n=2, expand=True)
country_codes_df.rename(columns={0: "country_code", 1: "state", 2: "city"}, inplace=True)
```

```python
country_codes_df.head()
```

```python
data['city'] = country_codes_df['city']
data['state'] = country_codes_df['state']
data['country_code'] = country_codes_df['country_code']
```

```python
data = data.merge(country_codes_df, on ="country_code")
data
```

```python
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline

sns.set_style('darkgrid')
matplotlib.rcParams['font.size'] = 16
matplotlib.rcParams['figure.figsize'] = (8, 4)
```

```python
pd.value_counts(data_df['fraudulent']).plot.bar()
```

```python
data_df.groupby(['fraudulent', 'industry']).mean()
```

```python
fake_jobs = data.fraudulent.value_counts()
fake_jobs
```

```python
plt.figure(figsize=(12,6))
plt.xticks(rotation=75)
plt.title('Job Postings')
sns.barplot(fake_jobs.index, fake_jobs);
plt.xlabel('Real=0 vs Fake=1')
plt.ylabel('Total NO of Jobs');
```

```python
topfake_job_listings = data.groupby('city_x')[['fraudulent']].sum()
topfake_job_listings = topfake_job_listings.sort_values('fraudulent',ascending=False).head(10)
topfake_job_listings
```

```python
plt.figure(figsize=(12,6))
plt.xticks(rotation=75)
plt.title('Top 10 states with fake job postings')
sns.barplot(topfake_job_listings.index, topfake_job_listings.fraudulent);
plt.xlabel('States with fake job postings')
plt.ylabel('Total NO of jobs');
```
